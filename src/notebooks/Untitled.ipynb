```python
from collections import Counter
import json
```

```python
#-*- coding:utf-8 -*-
```

```python
DATA = "../data/test.json"
```

```python
with open(DATA, "r") as file:
    data = json.load(file)
    print(len(data))
```

```python
parsed_data = []
for i, json_obj in enumerate(data):
    print("{} from {}".format(i, len(data)), end="\r")
    parsed_data.append(json_obj["ementa"])
```

```python
words_total = Counter()
for sentence in parsed_data:
    for word in sentence.split(" "):
        words_total.update([word])
```

```python
words_total.most_common()
```

```python
words = [word for word in words_total.items() if len(word[0]) > 3]
words.sort(key=lambda tup: tup[1])
words[-25:]
```
